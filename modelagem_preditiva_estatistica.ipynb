```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns # For enhanced visualizations
from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
```

```python
df = pd.read_csv('base_projeto_analise_quantitativa.csv')
print(df.head())
print(df.info())
print(df.describe())
#tipos de dados
print(df.dtypes)
```

```python
grupo1 = df[df['Resultado']==1]['Nivel_Satisfacao']
grupo0 = df[df['Resultado']==0]['Nivel_Satisfacao']

stats.ttest_ind(grupo1, grupo0)
X = df[['Idade', 'Salario', 'Nivel_Satisfacao']]
y = df['Resultado']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)   
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred_class = [1 if prob >= 0.5 else 0 for prob in y_pred]
print(classification_report(y_test, y_pred_class))  
```
